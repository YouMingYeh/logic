```python
import json
import sys

if len(sys.argv) != 2:
    print("Usage: {} <api.json>".format(sys.argv[0]))
    sys.exit(1)

with open("./api.json", "r") as f:
    api = json.load(f)

js_object = {}
for icon in api.get("icons", []):
    if "tags" in icon and icon["tags"]:
        tag = icon["tags"][0].replace(":", "")
        codepoint = icon.get("codepoint", "")
        js_object[tag] = "https://fonts.gstatic.com/s/e/notoemoji/latest/" + codepoint + "/lottie.json"

print("const api = {};".format(json.dumps(js_object, indent=2)))
```
